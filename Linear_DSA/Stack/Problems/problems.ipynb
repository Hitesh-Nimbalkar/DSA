```python
def nearest_greater_to_right(arr):
    result = [-1] * len(arr)
    stack = []

    for i in range(len(arr)-1, -1, -1):
        print(f" Loop for -{arr[i]}")
        while stack and arr[i] >= arr[stack[-1]]:
            popped_index = stack.pop()
            print(f"Popped: {arr[popped_index]} because {arr[i]} is greater")

        if stack:
            result[i] = arr[stack[-1]]
            print("---****----")
            print(result)
            print("---****----")
        
        
        
        print(f"Pushed: {arr[i]} ")
        stack.append(i)
        
        print("---------------------")

    return result

# Example usage:
input_array = [4, 5, 2, 10, 8]
result_array = nearest_greater_to_right(input_array)
print("Input Array:", input_array)
print("Nearest Greater to Right:", result_array)
```
