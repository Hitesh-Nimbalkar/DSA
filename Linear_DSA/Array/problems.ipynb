### Swap Sort 

1 to N Elements FInd Duplicate and Missing Elements 

O(n) way to Find 

Duplicate Found

```python
def find_duplicate_and_missing(arr):
    n = len(arr)
    duplicate = None
    missing = None

    i = 0
    while i < n:
        if arr[i] != i + 1:
            if arr[i] != arr[arr[i] - 1]:
                # Swap the elements
                arr[arr[i] - 1], arr[i] = arr[i], arr[arr[i] - 1]
            else:
                # Found a duplicate
                duplicate = arr[i]
                i += 1
        else:
            i += 1

    for i in range(n):
        if arr[i] != i + 1:
            # Found the missing number
            missing = i + 1
            break

    return duplicate, missing

# Example usage
arr = [4, 3, 2, 7, 8, 2, 1, 5]
duplicate, missing = find_duplicate_and_missing(arr)
print(f'Duplicate: {duplicate}, Missing: {missing}')
```
